```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#Load and View Data
```

```python
cereal_data = pd.read_csv('cereal.csv')
```

```python
cereal_data.info()
```

```python
print(cereal_data)
```

```python
tencal= cereal_data['calories'].nlargest(n=10)
print(tencal)
```

```python
tensug= cereal_data['sugars'].nlargest(n=10)
print(tensug)
```

```python
calsorted = cereal_data.sort_values(by='calories', ascending=False)
topcal = calsorted.head(10)
plt.figure(figsize=(10, 6))
plt.bar(topcal['name'], topcal['calories'])
plt.xlabel('Name')
plt.ylabel('Calories')
plt.title('Top 10 Calorie Values')
plt.xticks(rotation=45)  
plt.show()
```

```python
sugsorted = cereal_data.sort_values(by='sugars', ascending=False)
topsug = sugsorted.head(10)
plt.figure(figsize=(10, 6))
plt.bar(topsug['name'], topsug['sugars'])
plt.xlabel('Name')
plt.ylabel('sugars')
plt.title('Top 10 Sugar Values')
plt.xticks(rotation=45)  
plt.show()
```

```python
fibersorted = cereal_data.sort_values(by='fiber', ascending=True)
topfiber = fibersorted.head(10)
plt.figure(figsize=(10, 6))
plt.bar(topfiber['name'], topfiber['fiber'])
plt.xlabel('Name')
plt.ylabel('Fiber')
plt.title('Bottom 10 Fiber Values')
plt.xticks(rotation=45)  
plt.show()
```

```python
fibersorted1 = cereal_data.sort_values(by='fiber', ascending=False)
topfiber1 = fibersorted1.head(10)
plt.figure(figsize=(10, 6))
plt.bar(topfiber1['name'], topfiber1['fiber'])
plt.xlabel('Name')
plt.ylabel('Fiber')
plt.title('Top 10 Fiber Values')
plt.xticks(rotation=45)  
plt.show()
```

```python
ratingsorted = cereal_data.sort_values(by='rating', ascending=False)
toprating = ratingsorted.head(10)
plt.figure(figsize=(10, 6))
plt.bar(toprating['name'], toprating['rating'])
plt.xlabel('Name')
plt.ylabel('Rating')
plt.title('Top 10 Rating Values')
plt.xticks(rotation=45)  
plt.show()
```

```python
ratingsorted1 = cereal_data.sort_values(by='rating', ascending=True)
toprating1 = ratingsorted1.head(10)
plt.figure(figsize=(10, 6))
plt.bar(toprating1['name'], toprating1['rating'])
plt.xlabel('Name')
plt.ylabel('Rating')
plt.title('Top 10 Rating Values')
plt.xticks(rotation=45)  
plt.show()
```

```python
#Summary: After viewing various bar charts of sample data, I can conclude that fiber is a major factor in the cereals ratings.
# 1/2 of the bottom 10 in 'Fiber' are also in the bottom 10 in the total 'rating'.
# 1/2 of the top 10 in 'Fiber' are also in the top 10 in total 'rating'.
# Lastly 40% of the cereals in the top 10 in total 'calories' and 'sugar' are also in bottom 10 of 'rating'.
```
